```python
test_image_path = "../public/data/initial_test/original/"
fname = "479496.jpg"
```

```python
mimg = PImage.open(test_image_path + fname)
image_width, image_height = mimg.size

print(image_height, image_width)
```

```python
display(mimg)
```

```python
def make_bw(r,g,b):
    # double divide rounds to integer
    return (r + g + b) // 3 
```

```python
bwpxs = []
for r,g,b in get_pixels(mimg):
  gval = make_bw(r,g,b)
  bwpxs.append(gval)
```

```python
bw_img = make_image(bwpxs, mimg.size[0])
display(bw_img)
```

```python
from PIL import Image, ImageEnhance

# Open image
img = Image.open(test_image_path + fname).convert("L")  # Ensure it's grayscale

# Increase contrast
enhancer = ImageEnhance.Contrast(img)
enhanced_img = enhancer.enhance(2.0)  # 1.0 = original, >1 = more contrast
```
